## Objective
The objective of this project is to analyze student mental health data and identify patterns related to stress levels,sleep quality, study hours,
heart rate, and other lifestyle factors using Exploratory Data Analysis (EDA).

checking  data types

```python
df.dtypes
```

converting data types

```python
df["Step_Count"] = df["Step_Count"].astype(int)
```

```python
df["Stress_Levels"] = df["Stress_Levels"].astype(int)
```

```python
cols =[
    "Daily_Activity_Pattern",
    "Study_Hours",
    "Sleep_Schedules",
    "Class_Attendance",
    "Heart_Rate",
    "Sleep_Quality",
    "Messaging_Frequency",
    "Social_Media_Activity"
]
df[cols] = df[cols].round(2)
df
```

Outliers

```python
plt.figure(figsize=(10,5))

sns.boxplot(data=df[['Study_Hours',
                     'Heart_Rate',
                     'Sleep_Quality']])

plt.title("Outlier Detection")
plt.show()
```

```python
# Insight
• Study Hours contains a few outliers.
• Heart Rate has several extreme values.
• Sleep Quality has comparatively fewer outliers.
```

# Visualization

Distribution of Study Hours

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Study_Hours'], bins=20, kde=True)
plt.title("Distribution of Study Hours")
plt.xlabel("Study Hours")
plt.ylabel("Number of Students")
plt.show()
```

```python
# Insight
Most students study within a moderate range.
Very few students study extremely low or high hours.
The distribution is approximately normal.
```

Distribution of Heart Rate

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Heart_Rate'], bins=20)
plt.title("Distribution of Heart Rate")
plt.xlabel("Heart Rate")
plt.ylabel("Number of Students")
plt.show()
```

```python
# Insight
Heart rate values are concentrated around the average.
A few unusually high values indicate possible outliers
```

Distribution of Sleep Quality

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Sleep_Quality'], bins=20, kde=True)
plt.title("Distribution of Sleep Quality")
plt.xlabel("Sleep Quality")
plt.ylabel("Number of Students")
plt.show()
```

```python
# Insight
Most students report average sleep quality.
Very poor and excellent sleep quality are less common.
```

Study Hours vs Stress Levels

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='Study_Hours',y='Stress_Levels',data=df
)
plt.title("Study Hours vs Stress Levels")
plt.show()
```

```python
# Insight
Students with similar study hours show different stress levels.
No strong linear relationship is observed.
```

Sleep Quality vs Study Hours

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='Study_Hours',
    y='Sleep_Quality',
    alpha=0.7,
    data=df
)
plt.title("Study Hours vs Sleep Quality")
plt.show()
```

```python
# Insight
Students with longer study hours exhibit varying sleep quality.
No strong relationship is evident.
```
